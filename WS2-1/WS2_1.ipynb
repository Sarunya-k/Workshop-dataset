```python
bill_length_mm = penguin["bill_length_mm"].mean()
bill_depth_mm = penguin["bill_depth_mm"].mean()
flipper_length_mm = penguin["flipper_length_mm"].mean()
body_mass_g = penguin["body_mass_g"].mean()
body_mass_kg = penguin["body_mass_g"].mean()



penguin["species"].fillna("unknown", inplace = True)
penguin["island"].fillna("unknown", inplace = True)
penguin["bill_length_mm"].fillna(bill_length_mm, inplace = True)
penguin["bill_depth_mm"].fillna(bill_depth_mm, inplace = True)
penguin["flipper_length_mm"].fillna(flipper_length_mm, inplace = True)
penguin["body_mass_g"].fillna(body_mass_g, inplace = True)
penguin["body_mass_g"].fillna(body_mass_kg, inplace = True)

penguin = penguin.dropna()

print(penguin.to_string())
```
